### Filter with more than one condition


```python
df = df_orig
df.head()
```

```python
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 & mask2].head()
```

```python
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 | mask2].head()
```

### The `.isin()` Method


```python
df = df_orig
df.head()
```

```python
mask1 = df["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3]
```

```python
mask = df["Team"].isin(["Legal", "Sales", "Product"])
df[mask]
```

### The `.duplicated()` methods


```python
df = df_orig
df.sort_values(by="First Name", inplace=True)
df.head(15)
```

```python
one = df["First Name"].duplicated()
two = df["First Name"].duplicated(keep='last')
three = df["First Name"].duplicated(keep=False)
four = ~df["First Name"].duplicated(keep=False)

df[four]
```

### The `.drop_duplicates` method


```python
df = df_orig
df.sort_values(by="First Name", inplace=True)
df.head(5)
```

```python
len(df)
```

```python
len(df.drop_duplicates())
```

```python
len(df.drop_duplicates(subset=["First Name"], keep='last'))
len(df.drop_duplicates(subset=["First Name"], keep=False))
```

```python
df.drop_duplicates(subset=["Team"], keep=False)
```

```python
df.drop_duplicates(subset=["First Name", "Team"])
```
