### Rank values with the Rank method


```python
nba = nba_drop_orig
nba["Salary"] = nba["Salary"].fillna(0).astype('int')
nba.head(3)
```

```python
nba["rank"] = nba["Salary"].rank()
nba = nba.sort_values(by="Salary", ascending=False)
nba.head()
```

```python
data = {
    'Name': ['John', 'Emily', 'Charlie', 'Sophia', 'Jacob'],
    'Age': [25, 28, 32, 29, 35],
    'Country': ['USA', 'Canada', 'UK', 'Australia', 'Germany'],
    'Salary': [50000, 0, 0, 0, 70000]
}
df = pd.DataFrame(data)
df
```

```python
df["Rank"] = df["Salary"].rank()
df.sort_values(by="Salary", inplace=True)
df
```
