## Create New Column from Existing Column


```python
nba = pd.read_csv("./resources/nba.csv")
```

```python
nba["Age in a decade"] = nba["Age"] + 10
nba["Age in a decade"] = nba["Age"].add(10)
nba["Salary"] = nba["Salary"].sub(5000000)
nba["Weight in kgs"] = nba["Weight"].mul(0.4535)
nba.head()
```

## `.value_counts()` Method


```python
nba = pd.read_csv("./resources/nba.csv")
nba.head()
```

```python
nba["Position"].value_counts(normalize=True) * 100
nba["Salary"].value_counts()
```

## Drop Rows with Null Values


```python
nba.dropna(how="any")
nba.dropna(subset=["College", "Salary"])
# By default dropna() will remove rows and wont change the indexes
```

## Fill in Missing Values


```python
nba = pd.read_csv("./resources/nba.csv")
```

```python
nba.fillna(0)
nba["College"].fillna("Unknown", inplace=True)
nba["College"].tail()
nba["College"].copy().fillna("Unknown", inplace=True)
nba["Salary"] = nba["Salary"].fillna(0, inplace=True)
```

## The 'astype' Method 1


```python
nba = pd.read_csv("./resources/nba.csv").dropna(how="all")
```

```python
nba["Age"].hasnans
nba["Age"].astype("int")
nba["Age"] = nba["Age"].astype("int")
nba["Salary"].hasnans
# nba["Salary"].astype("int") -> Error
nba["Salary"] = nba["Salary"].fillna(0)
nba["Salary"].hasnans
nba["Salary"].astype("int")
```
